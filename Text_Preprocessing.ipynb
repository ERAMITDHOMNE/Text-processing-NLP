```python
text = """
Natural Language Processing (NLP) is a subfield of Artificial Intelligence (AI).
It focusses on the interaction between computers and human language.
NLP techniques includes tokenization, stemming, lemmatization, and named entity recognition (NER).
NLP helps machines understand and process human language data.
"""
```

# Tokenization

Breaking down of a sequence of text into sentences, words, or characters.

```python
import nltk
```

```python
from nltk.tokenize import sent_tokenize

tokenized_sentences = sent_tokenize(text)
tokenized_sentences
```

```python
from nltk.tokenize import word_tokenize

tokenized_words = [word_tokenize(sent) for
                   sent in tokenized_sentences]
tokenized_words
```

# Non Alphabets Removal

```python
alphabetic_words = [[word.lower() for word in sent if
                     word.isalpha()] for sent in
                    tokenized_words]
alphabetic_words
```

# Stopwords Removal

Filter out the common words that occur frequently in the text but typically carry little semantic meaning

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
stop_words
```

```python
filtered_words = [[word for word in sent if word not in
                   stop_words] for sent in alphabetic_words]
filtered_words
```

# Stemming and Lemmatization

Stemming and Lemmatization reduce words to their base or root form

Stemming is the process of removing suffixes or prefixes from a word to obtain its root form

Stemming is less computationally expensive lemmatization

Lemmatization is the process of converting a word to its base or root form (lemma)

Lemmatization requires access to a dictionary that maps the word to their lemmas

```python
flat_list = [word for sent in filtered_words for word in sent]
flat_list
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in flat_list]
stemmed_words
```

```python
nltk.download('omw-1.4')
# Multilingual Wordnet Data
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in
                    flat_list]
lemmatized_words
```

# Bag of Words

Converts unstructured text data into a numerical format that can be used by machine learning algorithms for various NLP tasks

Disregards the order of words in the text and only considers the frequency of words in the document

```python
from nltk import FreqDist

word_freq = FreqDist(flat_list)
word_freq
```

# Parts of Speech Tagging (POS Tagging)

Assigns grammatical labels (tags) to each word in a given text indicating its syntactic category or part of speech in the sentence

The: DT quick: JJ brown: JJ fox: NN jumps: VBZ over: IN the: DT lazy: JJ dog: NN

```python
from nltk import pos_tag

pos_tags = [pos_tag(sent) for sent in filtered_words]
pos_tags
```

# Named Entity Recognition

Extract and understand important entities within the text

Recognizes their semantic roles in the context

Organization: Apple Inc.
